```
df = pd.read_csv('anime.csv')
```

```
def extract_episode(txt):
    data = ""
    check = False
    for i in txt:
        if i == ')':
            check = False
            return data
        if check == True:
                data += i
        if i == '(':
            check = True
    return data
```

```
df['episode'] = df['Title'].apply(extract_episode)
df['episode'] = df['episode'].str.replace(' eps', '')
df
```

```
def extract_time(txt):
    date = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+1+19):
                    date += txt[j]


    return date

df['Title'].apply(extract_time)
```
