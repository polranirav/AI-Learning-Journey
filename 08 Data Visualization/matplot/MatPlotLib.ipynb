```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0, 5, 11)
x
y = x ** 2
y
```

```
plt.plot(x, y)
```

```
plt.subplot(2,2, 1)
plt.plot(x, y, 'r')
plt.subplot(2,2, 2)
plt.plot(y, x, 'g')
plt.subplot(2,2, 3)
plt.plot(x, y, 'b')
plt.subplot(2,2, 4)
plt.plot(y, x, 'y')
```

```
figure = plt.figure (figsize=(10, 5), dpi = 130)
axis_1 = figure.add_axes([0.9,0.9, 1,1])
axis_1.plot(x, y, 'r')
axis_2 = figure.add_axes([0.6, 0.6, 0.8, 0.8])
axis_2.plot(x, y, 'g')
```

```
plt.scatter(x, y, color='red', marker='o', s=100)
```

```
plt.hist(x, bins=10, color='blue', alpha=0.9)
```

```
plt.boxplot(x, vert=False, patch_artist=True, notch=True, boxprops=dict(facecolor='cyan', color='black'), medianprops=dict(color='red'))
```

```
figure = plt.figure (figsize=(10, 5), dpi = 130)
axis_1 = figure.add_axes([0.9,0.9, 1,1])
axis_1.plot(x, y, 'r')
plt.savefig('plot.png')
```
