```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = sns.load_dataset("flights")
df1
```

```python
df = sns.load_dataset('tips')
df
```

```python
tipscorr = df[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black', cbar_kws={"shrink": .8})
```

```python
sns.clustermap(tipscorr.corr())
```
