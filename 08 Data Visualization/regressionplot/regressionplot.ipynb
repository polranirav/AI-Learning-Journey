```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
tips = sns.load_dataset('tips')
```

```
tips
```

```
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, order=2, ci=None, palette='coolwarm',markers=['o','x'], scatter_kws={'s':100, 'alpha':0.5}, line_kws={'linewidth':2})
```
