```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.txt',sep = ';',header = None,names = ['text','emotion'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
unique_emotions = df['emotion'].unique()
emotion_number = {}
i = 0
for emotion in unique_emotions:
    emotion_number[emotion] = i
    i += 1

df['emotion'] = df['emotion'].map(emotion_number)
```

```python
df
```

```python
df['text'] = df['text'].map(lambda x: x.lower())
```

```python
df
```

```python
import string
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
df['text'] = df['text'].map(remove_punctuation)
```

```python
df
```

```python
def remove_numbers(text):
    new = ""
    for i in text:
        if not i.isdigit():
            new += i
    return new
```

```python
df['text'] = df['text'].map(remove_numbers)
df
```

```python
def remove_urls(text):
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
df['text'] = df['text'].map(remove_urls)
```

```python
def remove_emoji(text):
    new = ""
    for i in text:
        if i.isascii():
            new += i
    return new
```

```python
df['text'] = df['text'].map(remove_emoji)
```

```python
df
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
len(stop_words)
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
    words = txt.split()
    cleaner = []
    for word in words:
        if word not in stop_words:
            cleaner.append(word)
    return ' '.join(cleaner)

df['text'] = df['text'].apply(remove)
    
```

```python
df.loc[1]['text']
```

```python
def remove_wordtoken(txt):
    words = word_tokenize(txt)
    cleaner = []
    for word in words:
        if word not in stop_words:
            cleaner.append(word)
    return ' '.join(cleaner)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

bow_vectorizer = CountVectorizer()

X_train_bow = bow_vectorizer.fit_transform(X_train)
X_train_bow
```

```python
X_test_bow = bow_vectorizer.transform(X_test)
X_test_bow
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)
```

```python
predictions = nb_model.predict(X_test_bow)
```

```python
accuracy_score(y_test, predictions)
```

```python
tfidf_vectorizer = TfidfVectorizer()

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf, y_train)
```

```python
predict = nb_model.predict(X_test_tfidf)
predict
```

```python
accuracy_score(y_test, predict)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_tfidf, y_train)
```

```python
log_predictions = log_model.predict(X_test_tfidf)
```

```python
accuracy_score(y_test, log_predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train_tfidf, y_train)
```

```python
rf_predictions = rf_model.predict(X_test_tfidf)
rf_predictions
```

```python
accuracy_score(y_test, rf_predictions)
```
