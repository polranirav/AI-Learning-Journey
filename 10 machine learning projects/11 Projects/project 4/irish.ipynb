```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.species.unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('species', axis=1)
y = df['species']
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(train_x, train_y)
```

```python
model_knn.score(test_x, test_y)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(C = 10, kernel='linear',gamma = 'auto')
```

```python
model_svm.fit(train_x, train_y)
```

```python
model_svm.score(test_x, test_y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifire = GridSearchCV((model_svm),{
            'C' : [1,10,20,30],
            'kernel': ['linear','rbf','sigmoid'],
    },cv=5,return_train_score=False)
```

```python
classifire.fit(x,y)
```

```python
classifire.cv_results_
```

```python
result = pd.DataFrame(classifire.cv_results_)
```

```python
result.head()
```

```python
result[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifire = RandomizedSearchCV((model_svm),{
            'C' : [1,10,20,30],
            'kernel': ['linear','rbf','sigmoid'],
    },n_iter=30,cv=5,return_train_score=False)
```

```python
classifire.fit(x,y)
```

```python
classifire.cv_results_
result = pd.DataFrame(classifire.cv_results_)
result.head()
```

```python
result[['param_C','param_kernel','mean_test_score']]
```
